```python
#Feature Engineering:
df['accident_date']= pd.to_datetime(df['accident_date'],dayfirst=True)
```

### Question 1 - Current Year Casualties:

```python
#Year:
df['Year']=df['accident_date'].dt.year
```

```python
df['Year'].value_counts().to_frame()
```

```python
#Current_Year:
current_year=df['Year'].max()
current_year

#Previous Year:
previous_year=df['Year'].min()
previous_year

# current Year Casualties:
cy_casualties=df[df['Year']==current_year]['number_of_casualties'].sum()
print(f"Total Current Year Casualties:{cy_casualties}")

#Previous Year Casualties:
py_casualties=df[df['Year']==previous_year]['number_of_casualties'].sum()
print(f"Total Previous Year Casualties:{py_casualties}")
```

## cy-Fatal Casualties-2022

```python
df.accident_severity.value_counts()
```

```python
#Current Year Fatal Casualties-2022
cy_fatal_casualties=df[(df['Year']==current_year)&(df['accident_severity'].str.lower()=='fatal')]['number_of_casualties'].sum()

#Previous Year Fatal Casualties-2021
py_fatal_casualties=df[(df['Year']==previous_year)&(df['accident_severity'].str.lower()=='fatal')]['number_of_casualties'].sum()

print(f"Current Year Fatal Casualties:{cy_fatal_casualties}")
print(f"Previous Year Fatal Casualties:{py_fatal_casualties}")
```
